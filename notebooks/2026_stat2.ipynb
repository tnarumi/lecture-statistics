## 7. 提出に向けて

次の手順で提出前確認を行なってください．

- **メニューの「ファイル」から「保存」を選ぶことで，このNotebookを保存**
- **"全てを実行"** により以下のコードを実行
- 表示されるメッセージを確認し，必要に応じて内容を修正する

```python
from datetime import datetime, timezone, timedelta
from pathlib import Path
import re

SUBMISSION_FOLDER_NAME = "free-analysis"

jst = timezone(timedelta(hours=9))
now = datetime.now(jst)
errors = []
warnings = []


def get_text(name):
    return str(globals().get(name, "")).strip()


def is_blank_or_placeholder(value, placeholder_values):
    return (not value) or value in placeholder_values


def is_inside(child, parent):
    child = Path(child).resolve()
    parent = Path(parent).resolve()
    try:
        child.relative_to(parent)
        return True
    except ValueError:
        return False


def collect_filled_notices(prefix, placeholder_pattern):
    entries = []
    for name, value in sorted(globals().items()):
        if not name.startswith(prefix):
            continue
        text = str(value).strip()
        if not text:
            continue
        if re.fullmatch(placeholder_pattern, text):
            continue
        entries.append((name, text))
    return entries


# Google Colabでは，提出フォルダをGoogle Drive上に固定します．
try:
    from google.colab import drive
    drive.mount("/content/drive")
    submission_dir = Path("/content/drive/MyDrive/Colab Notebooks") / SUBMISSION_FOLDER_NAME
except ImportError:
    # Colab以外で確認する場合は，現在のフォルダまたは直下の free-analysis を提出フォルダとして扱います．
    cwd = Path.cwd()
    submission_dir = cwd if cwd.name == SUBMISSION_FOLDER_NAME else cwd / SUBMISSION_FOLDER_NAME
    if not submission_dir.exists():
        submission_dir = cwd

student_id = get_text("student_id")
student_name = get_text("student_name")
analysis_theme = get_text("analysis_theme")
stat_name = get_text("stat_name")
stat_title = get_text("stat_title")
stat_url = get_text("stat_url")
stat_date = get_text("stat_date")
csv_file = get_text("csv_file")

required_items = [
    ("student_id", student_id, ["学籍番号"], "学籍番号を入力してください．"),
    ("student_name", student_name, ["氏名"], "氏名を入力してください．"),
    ("analysis_theme", analysis_theme, ["分析テーマ"], "分析テーマを入力してください．"),
    ("stat_name", stat_name, ["統計名"], "統計名を入力してください．"),
    ("stat_title", stat_title, ["表題"], "表題を入力してください．"),
    ("stat_url", stat_url, ["https://www.e-stat.go.jp/dbview?sid=XXXXXXXX", "データ取得先URL"], "データ取得先URLを入力してください．"),
    ("stat_date", stat_date, ["20YY-MM-DD HH:MM", "公開年月日"], "公開年月日を入力してください．"),
    ("csv_file", csv_file, ["あなたが用意したデータファイル名.csv"], "csv_file にCSVファイル名を入力してください．"),
]

for name, value, placeholder_values, message in required_items:
    if is_blank_or_placeholder(value, placeholder_values):
        errors.append(f"{name}: {message}")

if stat_url and not is_blank_or_placeholder(stat_url, ["https://www.e-stat.go.jp/dbview?sid=XXXXXXXX", "データ取得先URL"]):
    if not stat_url.startswith(("http://", "https://")):
        errors.append("stat_url: URLは http:// または https:// から始まる形で入力してください．")

if stat_date and not is_blank_or_placeholder(stat_date, ["20YY-MM-DD HH:MM", "公開年月日"]):
    if not re.search(r"\d{4}[-/]\d{1,2}[-/]\d{1,2}", stat_date):
        warnings.append("stat_date は日付が分かる形式で入力してください．例: 2026-07-07 17:14")

print("提出前チェック")
print("現在時刻:", now.strftime("%Y-%m-%d %H:%M:%S JST"))
print("提出フォルダ:", submission_dir)
print()

print("入力内容")
print("- 学籍番号:", student_id or "未入力")
print("- 氏名:", student_name or "未入力")
print("- 分析テーマ:", analysis_theme or "未入力")
print("- 統計名:", stat_name or "未入力")
print("- 表題:", stat_title or "未入力")
print("- データ取得先URL:", stat_url or "未入力")
print("- 公開年月日:", stat_date or "未入力")
print("- CSVファイル:", csv_file or "未入力")
print()

csv_path = None
if not submission_dir.exists():
    errors.append(f"提出フォルダが見つかりません: {submission_dir}")
    errors.append(f"Google Drive の MyDrive/Colab Notebooks/ に {SUBMISSION_FOLDER_NAME} フォルダを作り，Notebookとデータを入れてください．")
else:
    if csv_file and not is_blank_or_placeholder(csv_file, ["あなたが用意したデータファイル名.csv"]):
        raw_csv_path = Path(csv_file).expanduser()
        if raw_csv_path.is_absolute():
            csv_path = raw_csv_path
            if not is_inside(csv_path, submission_dir):
                errors.append("csv_file には提出フォルダ内のCSVファイルを指定してください．提出フォルダ外のファイルは提出に含まれません．")
        else:
            csv_path = submission_dir / raw_csv_path

        if csv_path is not None:
            if not csv_path.exists():
                errors.append(f"CSVファイルが見つかりません: {csv_path}")
            elif csv_path.is_dir():
                errors.append(f"csv_file にはフォルダではなくCSVファイルを指定してください: {csv_path}")
            else:
                csv_label = csv_path.relative_to(submission_dir) if is_inside(csv_path, submission_dir) else csv_path
                print("CSVファイル確認:", csv_label, f"({csv_path.stat().st_size / 1024:.1f} KB)")
                if csv_path.suffix.lower() != ".csv":
                    warnings.append("csv_file で指定されたファイルの拡張子が .csv ではありません．CSVファイルを指定しているか確認してください．")

print()
if "df" not in globals():
    errors.append("df が見つかりません．データ読込セルを実行してください．")
else:
    try:
        print("データフレーム df の形:", df.shape)
        if df.empty or df.shape[1] == 0:
            errors.append("df が空です．csv_file や skiprows の指定を確認してください．")
        unnamed_columns = [str(col) for col in df.columns if str(col).startswith("Unnamed")]
        if unnamed_columns:
            warnings.append("列名に Unnamed が含まれています．e-Statのメタデータ行が列名として読み込まれていないか，skiprows を確認してください．")
    except Exception as exc:
        errors.append(f"df の確認中にエラーが出ました: {exc}")

print()
print("データ確認メモの確認")
data_entries = collect_filled_notices("data_notice", r"気づいたこと\d*")
if data_entries:
    print(f"データ確認メモが {len(data_entries)} 件入力されています．")
    for name, text in data_entries:
        short_text = text if len(text) <= 80 else text[:77] + "..."
        print(f"- {name}: {short_text}")
else:
    errors.append("data_notice1 などに，データを確認して気づいたことを1つ以上入力してください．")

print()
print("考察の確認")
analysis_entries = collect_filled_notices("analysis_notice", r"分析結果や考察\d*")
if analysis_entries:
    print(f"考察が {len(analysis_entries)} 件入力されています．")
    for name, text in analysis_entries:
        short_text = text if len(text) <= 80 else text[:77] + "..."
        print(f"- {name}: {short_text}")
else:
    errors.append("analysis_notice1 などに，考察を1つ以上入力してください．")

print()
print("提出対象ファイル一覧")
submitted_files = []
notebook_files = []
if submission_dir.exists():
    for file_path in sorted(p for p in submission_dir.rglob("*") if p.is_file()):
        rel = file_path.relative_to(submission_dir)
        if any(part in {".ipynb_checkpoints", "__pycache__", "__MACOSX"} for part in rel.parts):
            continue
        if file_path.name in {".DS_Store"}:
            continue
        submitted_files.append(file_path)
        if file_path.suffix == ".ipynb":
            notebook_files.append(file_path)
        size_kb = file_path.stat().st_size / 1024
        print(f"- {rel} ({size_kb:.1f} KB)")

if submission_dir.exists() and not submitted_files:
    errors.append("提出フォルダ内に提出対象ファイルがありません．Notebookとデータを入れてください．")

if not notebook_files:
    errors.append("提出フォルダ内に .ipynb ファイルが見つかりません．Notebookを提出フォルダに入れてください．")

if csv_path is not None and csv_path.exists() and csv_path.is_file() and not is_inside(csv_path, submission_dir):
    errors.append("指定されたCSVファイルは提出フォルダ外にあります．提出フォルダ内のCSVファイルを指定してください．")

print()
if notebook_files:
    print("Notebookの保存時刻")
    newest_mtime = None
    for notebook_path in notebook_files:
        mtime = datetime.fromtimestamp(notebook_path.stat().st_mtime, jst)
        newest_mtime = mtime if newest_mtime is None or mtime > newest_mtime else newest_mtime
        age_minutes = (now - mtime).total_seconds() / 60
        print(f"- {notebook_path.relative_to(submission_dir)}: {mtime.strftime('%Y-%m-%d %H:%M:%S JST')}（約{age_minutes:.1f}分前）")

    newest_age_minutes = (now - newest_mtime).total_seconds() / 60
    if newest_age_minutes > 15:
        warnings.append("最新のNotebook保存時刻が15分以上前です．提出前にColabで ファイル > 保存，または Ctrl/Cmd + S を実行してください．")

if Path.cwd() != submission_dir and submission_dir.exists():
    warnings.append(f"現在の作業フォルダは提出フォルダではありません．CSVの読み込みで困る場合は，提出フォルダを作業フォルダにしてから分析してください．現在: {Path.cwd()}")

print()
if warnings:
    print("注意")
    for warning in warnings:
        print("-", warning)
    print()

if errors:
    print("修正が必要な項目")
    for error in errors:
        print("-", error)
    raise ValueError("提出前チェックで修正が必要な項目があります．上のメッセージを確認してください．")

print("提出前チェックは完了です．")
print(f"Google Driveで MyDrive/Colab Notebooks/{SUBMISSION_FOLDER_NAME} を右クリックし，ダウンロードしたZIPをMoodleに提出してください．")
```
